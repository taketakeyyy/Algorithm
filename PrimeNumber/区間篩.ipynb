```python
def segment_sieve(a, b):
    """区間 [a,b) に対してエラトステネスの篩をする
    
    Args:
        a(int): 正整数
        b(int): 正整数
    
    Returns:
        (tuple): p, primes
        p(int): 素数の数
        primes(list): 素数のリスト
    """
    import math
    # [a,b)の整数に対して篩をかける。
    # is_prime[i-a] = True ⇔ iが素数
    is_prime_small = [True for _ in range(math.ceil(b**(1/2)))]
    is_prime = [True for _ in range(b-a)]

    i = 2
    while i*i < b:
        if is_prime_small[i]:
            
            # [2, √b)の篩
            j = i*2
            while j*j < b:
                is_prime_small[j] = False
                j += i
            
            # [a, b)の篩
            j = max(2, (a+i-1)//i)*i
            print(f"{j}-{b}")
            while j < b:
                is_prime[j-a] = False
                j += i
        i += 1
    
    p, primes = 0, []
    for i, x in enumerate(is_prime):
        if x:
            print(i)
            p += 1
            primes.append(a+i)
    
    return p, primes
```

```python
p, primes = segment_sieve(6, 18)
print(p)
print(primes)
```
